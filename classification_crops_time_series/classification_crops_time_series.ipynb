## Выделим выборки

```python
X = df.drop(["id",".geo", "crop"], axis = 1)
y = df[["crop"]]
```

```python
test = test.drop(["id",".geo"], axis = 1)
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Обучение модели

```python
%%time
xgb_model = XGBClassifier(random_state=42)

parameters = [{'num_feature': [10, 30, 50, 71],
               'max_depth': [3, 6, 10, 20]}]

xgb_model_gscv = GridSearchCV(xgb_model, param_grid=parameters, scoring='roc_auc', 
                   cv=3, verbose=True, n_jobs=-1)
xgb_model_gscv.fit(X_train, y_train)

print(xgb_model_gscv.best_score_, xgb_model_gscv.best_params_)
```

```python
xgb_model = XGBClassifier(random_state=42, max_depth=3, num_feature=10)
```

```python
xgb_model.fit(X, y)
```

```python
%%time
lgbm_model = LGBMClassifier(random_state=42)

parameters = [{'boosting_type': ['gbdt', 'goss', 'dart'],
            'num_leaves': [2, 10, 50, 100],
            'learning_rate': [0.01,0.1,0.5],
            'colsample_bytree': [0.6,0.8,1],
            "max_depth": [2, 5, 7, 20]}]

lgbm_model_gscv = GridSearchCV(lgbm_model, param_grid=parameters, scoring='roc_auc', 
                   cv=3, verbose=True, n_jobs=-1)
lgbm_model_gscv.fit(X_train, y_train)

print(lgbm_model_gscv.best_score_, lgbm_model_gscv.best_params_)
```
